```
# Synthetic Data Generation
import pandas as pd
from ctgan import CTGAN
import warnings

warnings.filterwarnings('ignore')

print("--- Starting Synthetic Data Generation ---")

# Load the data
try:
    real_data = pd.read_csv(
        "/content/drive/MyDrive/Final Year Project/Updated_Prakriti_With_Features.csv",
        dtype=str        # OK only if dataset is fully categorical
    )
    print(f"\nSuccessfully loaded dataset with {len(real_data)} rows.")
except FileNotFoundError:
    print("CSV not found.")
    raise SystemExit

# Identify categorical columns (all columns are treated as categorical)
discrete_columns = real_data.columns.tolist()

print("\nDetected categorical columns:")
print(discrete_columns)

# Initialize CTGAN
synthesizer = CTGAN(
    epochs=300,
    batch_size=50,
    verbose=True
)

print("\nTraining the CTGAN model...")
synthesizer.fit(real_data, discrete_columns=discrete_columns)

print("\n--- Training Complete ---")

# Generate synthetic samples
num_new_rows = 5000
print(f"\nGenerating {num_new_rows} synthetic rows...")
synthetic_data = synthesizer.sample(num_new_rows)

# Save to a NEW FILE (VERY IMPORTANT)
output_filename = "/content/drive/MyDrive/Final Year Project/synthetic_prakriti_data_5000.csv"
synthetic_data.to_csv(output_filename, index=False)

print(f"\nSuccessfully saved synthetic data to '{output_filename}'.")
```
